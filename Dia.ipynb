```python
df['Class'] = df['Run'].apply(lambda x: 'Pool' if 'Pool' in x else 'Sample')
df['Batch'] = df['Run'].str.extract(r'(Batch-\d+)')
```

```python
pool_samples = df[df['Class'] == 'Pool']
```

```python
def plot_bar_with_values(dataframe, x_col, y_col):
    plt.figure(figsize=(8,4))
    ax = sns.barplot(x=x_col, y=y_col, data=dataframe, palette='viridis', alpha=0.7, edgecolor="black")
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2, height, f'{height:.0f}', ha='center', va='bottom')
    plt.show()
```

```python
pool_protein_counts = pool_samples.groupby(['Run']).agg({'Protein.Group':'nunique','Stripped.Sequence':'nunique','Precursor.Id':'nunique',
                                                        'Batch':'last'}).reset_index()
pool_protein_counts = pool_protein_counts.sort_values(by='Run')
```

```python
def plot_combined_catplot_swarmplot(data, x_value_box, y_value_box, x_value_swarm, y_value_swarm, height=5, aspect=3):
    g = sns.catplot(x=x_value_box, y=y_value_box, data=data, palette='colorblind', kind='box', height=height, aspect=aspect, 
                    hue='Batch', legend=False, dodge=False, showfliers=False, boxprops=dict(alpha=0.5))
    ax = g.ax
    sns.swarmplot(x=x_value_swarm, y=y_value_swarm, data=data, palette='colorblind', hue='Run', dodge=False, alpha=1, ax=ax, legend = False)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    legend_font_size = int(10 * aspect / height)
    handles, labels = ax.get_legend_handles_labels()
    unique_labels = dict(zip(labels, handles))
    plt.legend(unique_labels.values(), unique_labels.keys(), fontsize=legend_font_size, loc='upper right', bbox_to_anchor=(1.07, 1))
    plt.show()
```

```python
plot_combined_catplot_swarmplot(pool_protein_counts,'Batch','Protein.Group','Batch','Protein.Group')
plot_combined_catplot_swarmplot(pool_protein_counts,'Batch','Stripped.Sequence','Batch','Stripped.Sequence')
plot_combined_catplot_swarmplot(pool_protein_counts,'Batch','Precursor.Id','Batch','Precursor.Id')
```

## Calculating CV for full profile proteingroups in Pooled samples run in every batch (QC)

```python
pool_samples_pro_quant = pool_samples.drop_duplicates(['Run','Protein.Group'],keep='last')
pool_protein_cv_df = pool_samples_pro_quant.groupby(['Batch','Protein.Group']).agg({'PG.Quantity':['mean','std'],
                                                                                   'Run':'nunique'}).reset_index()
pool_protein_cv_df.columns = ['Batch', 'Protein.Group', 'Mean', 'Std','nunique']
pool_protein_cv_df2 = pool_protein_cv_df[pool_protein_cv_df['nunique'] == 4]
pool_protein_cv_df2['cv'] = pool_protein_cv_df2['Std']/pool_protein_cv_df2['Mean'] * 100
```

```python
sns.catplot(data=pool_protein_cv_df2,x='Batch',y='cv',hue='Batch',kind='violin',dodge=False,height=5,aspect=2, palette = 'colorblind')
plt.xticks(rotation=45,ha='right')
plt.ylim(0,100)
plt.show()
```

```python
cluster_map_df = pool_samples_pro_quant.pivot(index=['Protein.Group'],columns = 'Run', values = 'PG.Quantity')
cluster_map_df = np.log10(cluster_map_df)
cluster_map_df1 = cluster_map_df.dropna(axis=0)
cluster_map_df2 = cluster_map_df1.corr(method='spearman')
sns.clustermap(cluster_map_df2.corr(method='spearman'), cmap='viridis', col_cluster=True)
```

```python
from matplotlib.patches import Patch
run_batch_df = pool_samples_pro_quant[['Run', 'Batch']].drop_duplicates().set_index('Run')
unique_batches = run_batch_df['Batch'].unique()
palette = sns.color_palette('hsv', len(unique_batches))
batch_colors = dict(zip(unique_batches, palette))
batch_colors_mapped = run_batch_df['Batch'].map(batch_colors)
sns.clustermap(cluster_map_df2, cmap='viridis', col_cluster=True, row_cluster=True, col_colors=batch_colors_mapped)
legend_elements = [Patch(facecolor=batch_colors[batch], label=batch) for batch in unique_batches]
plt.legend(handles=legend_elements, title="Batch", bbox_to_anchor=(1.1, 1), bbox_transform=plt.gcf().transFigure)
plt.show()
```
